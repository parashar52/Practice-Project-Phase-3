# Modelling

Finding the best model

```python
#importing required libraries
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score,accuracy_score
from sklearn.linear_model import LinearRegression
from sklearn.naive_bayes import MultinomialNB
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=45)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
y_test.shape
```

# Linear Regression

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
predlr=lr.predict(x_test)
```

```python
r2_score(predlr,y_test)
```

```python
mean_absolute_error(predlr,y_test)
```

```python
mean_squared_error(predlr,y_test)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dtr=DecisionTreeRegressor()
dtr.fit(x_train,y_train)
```

```python
preddtr=dtr.predict(x_test)
```

```python
print('r2_score:',r2_score(preddtr,y_test))
print('Mean Absolute Error:',mean_absolute_error(preddtr,y_test))
print('Mean Squared Error:',mean_squared_error(preddtr,y_test))
```

# Saving the best model
Linear Regression with .92 is the best model of the two

```python
import joblib
joblib.dump(lr,'lrfile.obj')

lr_from_joblib=joblib.load('lrfile.obj')
lr_from_joblib.predict(x_test)
```
